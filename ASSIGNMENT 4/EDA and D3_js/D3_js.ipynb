```python
# Enhanced D3.js bar chart visualization code
d3_stylish_bar_chart_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Stylish D3 Bar Chart</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>
<div id="d3-container"></div>
<script>
    var data = """ + str(category_sales.to_dict(orient='records')) + """;
    var svgWidth = 600, svgHeight = 400, barPadding = 5;
    var barWidth = (svgWidth / data.length);
    var margin = {top: 20, right: 20, bottom: 50, left: 50};
    var width = svgWidth - margin.left - margin.right;
    var height = svgHeight - margin.top - margin.bottom;

    var svg = d3.select('#d3-container')
        .append('svg')
        .attr('width', svgWidth)
        .attr('height', svgHeight)
        .append('g')
        .attr('transform', 'translate(' + margin.left + ',' + margin.top + ')');

    var yScale = d3.scaleLinear()
        .domain([0, d3.max(data, d => d.Total_Sales)])
        .range([height, 0]);

    var xScale = d3.scaleBand()
        .domain(data.map(d => d.Category))
        .range([0, width])
        .padding(0.3);

    svg.selectAll('rect')
        .data(data)
        .enter()
        .append('rect')
        .attr('x', d => xScale(d.Category))
        .attr('y', d => yScale(d.Total_Sales))
        .attr('height', d => height - yScale(d.Total_Sales))
        .attr('width', xScale.bandwidth())
        .attr('fill', 'steelblue')
        .on('mouseover', function() {
            d3.select(this)
                .attr('fill', 'darkorange');
        })
        .on('mouseout', function() {
            d3.select(this)
                .attr('fill', 'steelblue');
        })
        .transition()
        .duration(1000)
        .attr('y', d => yScale(d.Total_Sales))
        .attr('height', d => height - yScale(d.Total_Sales));

    svg.append('g')
        .attr('transform', 'translate(0,' + height + ')')
        .call(d3.axisBottom(xScale))
        .selectAll('text')
        .style('text-anchor', 'end')
        .attr('dx', '-.8em')
        .attr('dy', '.15em')
        .attr('transform', 'rotate(-45)')
        .attr('fill', 'darkslategray');

    svg.append('g')
        .call(d3.axisLeft(yScale).ticks(5))
        .selectAll('text')
        .attr('fill', 'darkslategray');

    svg.append('text')
        .attr('x', width/2)
        .attr('y', height + margin.bottom - 5)
        .attr('text-anchor', 'middle')
        .attr('fill', 'darkslategray')
        .text('Product Category');

    svg.append('text')
        .attr('x', -height/2)
        .attr('y', -margin.left + 10)
        .attr('text-anchor', 'middle')
        .attr('fill', 'darkslategray')
        .attr('transform', 'rotate(-90)')
        .text('Total Sales');
</script>
</body>
</html>
"""

# Display the enhanced D3.js bar chart in the notebook
display(HTML(d3_stylish_bar_chart_code))
```

```python
# Define total_sales_data
total_sales_data = data['Total_Sales'].tolist()

# D3.js histogram visualization code
d3_histogram_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>D3 Histogram</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>
<div id="d3-container"></div>
<script>
    var data = """ + str(total_sales_data) + """;
    var svgWidth = 600, svgHeight = 300, barPadding = 1;

    var svg = d3.select('#d3-container')
        .append('svg')
        .attr('width', svgWidth)
        .attr('height', svgHeight);

    var histogram = d3.histogram()
        .domain([0, d3.max(data)])
        .thresholds(30);  // Number of bins

    var bins = histogram(data);

    var yScale = d3.scaleLinear()
        .domain([0, d3.max(bins, d => d.length)])
        .range([svgHeight, 0]);

    var xScale = d3.scaleLinear()
        .domain([0, d3.max(data)])
        .range([0, svgWidth]);

    svg.selectAll('rect')
        .data(bins)
        .enter()
        .append('rect')
        .attr('x', d => xScale(d.x0))
        .attr('y', d => yScale(d.length))
        .attr('width', d => xScale(d.x1) - xScale(d.x0) - barPadding)
        .attr('height', d => svgHeight - yScale(d.length))
        .attr('fill', '#69b3a2');
</script>
</body>
</html>
"""

# Display the D3.js histogram in the notebook
display(HTML(d3_histogram_code))
```

```python
# D3.js pie chart visualization code
d3_pie_chart_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>D3 Pie Chart</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>
<div id="d3-container"></div>
<script>
    var data = """ + str(category_sales.to_dict(orient='records')) + """;
    var svgWidth = 400, svgHeight = 400, radius = Math.min(svgWidth, svgHeight) / 2;

    var svg = d3.select('#d3-container')
        .append('svg')
        .attr('width', svgWidth)
        .attr('height', svgHeight)
        .append('g')
        .attr('transform', 'translate(' + (svgWidth / 2) + ',' + (svgHeight / 2) + ')');

    var color = d3.scaleOrdinal(d3.schemeCategory10);

    var pie = d3.pie().value(d => d.Total_Sales);
    var path = d3.arc().outerRadius(radius).innerRadius(0);

    var arc = svg.selectAll('arc')
        .data(pie(data))
        .enter()
        .append('g');

    arc.append('path')
        .attr('d', path)
        .attr('fill', (d, i) => color(i));

    arc.append('text')
        .attr('transform', d => 'translate(' + path.centroid(d) + ')')
        .attr('dy', '.35em')
        .style('text-anchor', 'middle')
        .text(d => d.data.Category);
</script>
</body>
</html>
"""

# Display the D3.js pie chart in the notebook
display(HTML(d3_pie_chart_code))
```

```python
# Group data by order date and sum the total sales for each date
time_series_data = data.groupby('Order_Date').Total_Sales.sum().reset_index()

# Sort the data by date
time_series_data = time_series_data.sort_values(by='Order_Date')

time_series_data.head()
```

```python
# D3.js line chart visualization code
d3_line_chart_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>D3 Line Chart</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>
<div id="d3-container"></div>
<script>
    var data = """ + str(time_series_data.to_dict(orient='records')) + """;
    var svgWidth = 700, svgHeight = 400, margin = {top: 20, right: 20, bottom: 30, left: 50};
    var width = svgWidth - margin.left - margin.right;
    var height = svgHeight - margin.top - margin.bottom;

    var svg = d3.select('#d3-container')
        .append('svg')
        .attr('width', svgWidth)
        .attr('height', svgHeight)
        .append('g')
        .attr('transform', 'translate(' + margin.left + ',' + margin.top + ')');

    var x = d3.scaleTime().rangeRound([0, width]);
    var y = d3.scaleLinear().rangeRound([height, 0]);

    var line = d3.line()
        .x(d => x(new Date(d.Order_Date)))
        .y(d => y(d.Total_Sales));

    x.domain(d3.extent(data, d => new Date(d.Order_Date)));
    y.domain(d3.extent(data, d => d.Total_Sales));

    svg.append('g')
        .attr('transform', 'translate(0,' + height + ')')
        .call(d3.axisBottom(x))
        .select(".domain")
        .remove();

    svg.append('g')
        .call(d3.axisLeft(y))
        .append('text')
        .attr('fill', '#000')
        .attr('transform', 'rotate(-90)')
        .attr('y', 6)
        .attr('dy', '0.71em')
        .attr('text-anchor', 'end')
        .text('Total Sales');

    svg.append('path')
        .datum(data)
        .attr('fill', 'none')
        .attr('stroke', 'steelblue')
        .attr('stroke-linejoin', 'round')
        .attr('stroke-linecap', 'round')
        .attr('stroke-width', 1.5)
        .attr('d', line);
</script>
</body>
</html>
"""

# Display the D3.js line chart in the notebook
display(HTML(d3_line_chart_code))
```

```python
# Use the original data to extract required columns for scatter plot
scatter_data = data[['Category', 'Quantity_Sold', 'Price_Per_Unit']]

scatter_data.head()
```

```python
# D3.js scatter plot visualization code
d3_scatter_plot_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>D3 Scatter Plot</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>
<div id="d3-container"></div>
<script>
    var data = """ + str(scatter_data.to_dict(orient='records')) + """;
    var svgWidth = 700, svgHeight = 500, margin = {top: 20, right: 20, bottom: 30, left: 50};
    var width = svgWidth - margin.left - margin.right;
    var height = svgHeight - margin.top - margin.bottom;

    var svg = d3.select('#d3-container')
        .append('svg')
        .attr('width', svgWidth)
        .attr('height', svgHeight)
        .append('g')
        .attr('transform', 'translate(' + margin.left + ',' + margin.top + ')');

    var x = d3.scaleLinear().range([0, width]);
    var y = d3.scaleLinear().range([height, 0]);

    var color = d3.scaleOrdinal(d3.schemeCategory10);

    x.domain([0, d3.max(data, d => d.Quantity_Sold)]);
    y.domain([0, d3.max(data, d => d.Price_Per_Unit)]);

    svg.selectAll('dot')
        .data(data)
        .enter().append('circle')
        .attr('r', 5)
        .attr('cx', d => x(d.Quantity_Sold))
        .attr('cy', d => y(d.Price_Per_Unit))
        .style('fill', d => color(d.Category));

    svg.append('g')
        .attr('transform', 'translate(0,' + height + ')')
        .call(d3.axisBottom(x));

    svg.append('g')
        .call(d3.axisLeft(y));

    var legend = svg.selectAll('.legend')
        .data(color.domain())
        .enter().append('g')
        .attr('class', 'legend')
        .attr('transform', (d, i) => 'translate(0,' + i * 20 + ')');

    legend.append('rect')
        .attr('x', width - 18)
        .attr('width', 18)
        .attr('height', 18)
        .style('fill', color);

    legend.append('text')
        .attr('x', width - 24)
        .attr('y', 9)
        .attr('dy', '.35em')
        .style('text-anchor', 'end')
        .text(d => d);
</script>
</body>
</html>
"""

# Display the D3.js scatter plot in the notebook
display(HTML(d3_scatter_plot_code))
```

```python
# Group data by order date and category, then sum the quantity sold for each group
stacked_data = data.groupby(['Order_Date', 'Category']).Quantity_Sold.sum().unstack().reset_index()
stacked_data.fillna(0, inplace=True)

# Sort the data by date
stacked_data = stacked_data.sort_values(by='Order_Date')

stacked_data.head()
```

```python
# D3.js stacked bar chart visualization code
d3_stacked_bar_chart_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>D3 Stacked Bar Chart</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>
<div id="d3-container"></div>
<script>
    var data = """ + str(stacked_data.to_dict(orient='records')) + """;
    var categories = ['Apparel', 'Electronics', 'Home and Living'];
    var svgWidth = 700, svgHeight = 500, margin = {top: 20, right: 20, bottom: 30, left: 50};
    var width = svgWidth - margin.left - margin.right;
    var height = svgHeight - margin.top - margin.bottom;

    var svg = d3.select('#d3-container')
        .append('svg')
        .attr('width', svgWidth)
        .attr('height', svgHeight)
        .append('g')
        .attr('transform', 'translate(' + margin.left + ',' + margin.top + ')');

    var x = d3.scaleBand().range([0, width]).padding(0.4);
    var y = d3.scaleLinear().range([height, 0]);
    var color = d3.scaleOrdinal(d3.schemeCategory10);

    x.domain(data.map(d => d.Order_Date));
    y.domain([0, d3.max(data, d => d.Apparel + d.Electronics + d['Home and Living'])]);

    svg.append('g')
        .attr('transform', 'translate(0,' + height + ')')
        .call(d3.axisBottom(x).tickSizeOuter(0));

    svg.append('g')
        .call(d3.axisLeft(y));

    var stack = d3.stack().keys(categories);
    var series = stack(data);

    svg.append('g')
        .selectAll('g')
        .data(series)
        .enter().append('g')
        .attr('fill', d => color(d.key))
        .selectAll('rect')
        .data(d => d)
        .enter().append('rect')
        .attr('x', d => x(d.data.Order_Date))
        .attr('y', d => y(d[1]))
        .attr('height', d => y(d[0]) - y(d[1]))
        .attr('width', x.bandwidth());

    var legend = svg.append('g')
        .selectAll('g')
        .data(categories.slice())
        .enter().append('g')
        .attr('transform', (d, i) => 'translate(0,' + i * 20 + ')');

    legend.append('rect')
        .attr('x', width - 19)
        .attr('width', 19)
        .attr('height', 19)
        .attr('fill', color);

    legend.append('text')
        .attr('x', width - 24)
        .attr('y', 9.5)
        .attr('dy', '0.32em')
        .text(d => d);
</script>
</body>
</html>
"""

# Display the D3.js stacked bar chart in the notebook
display(HTML(d3_stacked_bar_chart_code))
```
