```python
import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions

def parse_csv(line):
    # Split the line into a list
    fields = line.split(',')
    return (fields[2], 1)  # Return the country and a count of 1

# Define the pipeline
pipeline_options = PipelineOptions()
with beam.Pipeline(options=pipeline_options) as p:
    (p
     | "Read CSV" >> beam.io.ReadFromText('/content/car_manufacturers.csv', skip_header_lines=1)
     | "Parse CSV" >> beam.Map(parse_csv)
     | "Count by Country" >> beam.CombinePerKey(sum)
     | "Format Results" >> beam.Map(lambda result: "{},{}".format(result[0], result[1]))
     | "Write to Output CSV" >> beam.io.WriteToText('/content/output_counts_by_country.csv', header='Country,Count', shard_name_template=''))

```
